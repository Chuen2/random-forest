```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score
import time
```

```python
train_csv = pd.read_csv("C:/Users/TANG YOON CHUEN/Desktop/UMS_files/FYP\Dataset/archive/Human Action Recognition/Training_set.csv")
test_csv = pd.read_csv("C:/Users/TANG YOON CHUEN/Desktop/UMS_files/FYP/Dataset/archive/Human Action Recognition/Testing_set.csv")
```

```python
img_data = []
img_label = []
```

```python
for i in range(len(train_csv)):
    img = "C:/Users/TANG YOON CHUEN/Desktop/UMS_files/FYP/Dataset/archive/Human Action Recognition/train/" + train_csv["filename"][i]
    img_data.append(img)
    img_label.append(train_csv["label"][i])
```

```python
img_label = np.array(img_label)
```

```python
# Encode the labels
encoder = LabelEncoder()
img_label = encoder.fit_transform(img_label)
num_classes = len(encoder.classes_)
```

```python
# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(img_data, img_label, test_size=0.2, random_state=42)
```

```python
# Load and preprocess images (you can use a different image processing library if needed)
def load_and_preprocess_images(image_paths, target_size=(128, 128)):
    image_data = []
    for image_path in image_paths:
       
        from skimage.io import imread
        from skimage.transform import resize
        image = imread(image_path)
        image = resize(image, target_size)
        image_data.append(image)
    return np.array(image_data)
```

```python
X_train = load_and_preprocess_images(X_train)
X_val = load_and_preprocess_images(X_val)
```

```python
# Flatten the image data for Random Forest
X_train_flatten = X_train.reshape(X_train.shape[0], -1)
X_val_flatten = X_val.reshape(X_val.shape[0], -1)
```

```python
# Create and train a Random Forest Classifier with max_depth and min_samples_leaf set
rf_classifier = RandomForestClassifier(n_estimators=120, max_depth=10, min_samples_leaf=3, random_state=42)
```

```python
train_accuracy = [] 
val_accuracy = []
train_loss = [] 
val_loss = []
```

```python
start_time = time.time()
```

```python
for n_estimators in range(1, 121):  
    rf_classifier.set_params(n_estimators=n_estimators)
    rf_classifier.fit(X_train_flatten, y_train)

    y_train_pred = rf_classifier.predict(X_train_flatten)
    train_accuracy.append(accuracy_score(y_train, y_train_pred))

    y_val_pred = rf_classifier.predict(X_val_flatten)
    val_accuracy.append(accuracy_score(y_val, y_val_pred))

    train_loss.append(1 - accuracy_score(y_train, y_train_pred))
    val_loss.append(1 - accuracy_score(y_val, y_val_pred))
```

```python
end_time = time.time()
```

```python
total_time = end_time - start_time
```

```python
for n_estimators in range(1, 121):
    print(f"Number of Estimators: {n_estimators}")
    print(f"Train Accuracy: {train_accuracy[n_estimators - 1]:.4f}")
    print(f"Validation Accuracy: {val_accuracy[n_estimators - 1]:.4f}")
    print(f"Train Loss: {train_loss[n_estimators - 1]:.4f}")
    print(f"Validation Loss: {val_loss[n_estimators - 1]:.4f}")
    print()
```

```python
print(f"Total Time Taken: {total_time:.2f} seconds")
```
